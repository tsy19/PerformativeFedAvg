## Impact of batch size

```python
experiment_id = client.get_experiment_by_name('b').experiment_id
runs = mlflow.search_runs(experiment_id)
```

```python
subruns = runs
# with pd.option_context('display.max_columns', None):
#     display(subruns)
print(subruns.shape)

indices = {
    'Full participation': ['full', (subruns['params.K'] == '10')],
    'Scheme I': ['i', (subruns['params.K'] == '5') & (subruns['params.replace'] == 'True')],
    'Scheme II': ['ii', (subruns['params.K'] == '5') & (subruns['params.replace'] == 'False')],
}

for k, (name, idx) in indices.items():
    for query, ylabel in [
        ['theta_dist', 'distance to the stable solution'],
        ['loss', 'loss']
    ]: #, 'concensus_error']:
        plt.figure(figsize=(3.6,2.8))
        values = [[e.value for e in client.get_metric_history(run_id, query)]
                   for run_id in subruns.loc[idx, 'run_id']]
        steps = [[e.step for e in client.get_metric_history(run_id, query)]
                   for run_id in subruns.loc[idx, 'run_id']]
        labels = subruns.loc[idx, 'params.batch_size']
        # breakpoint()
        for s, v, l in zip(steps, values, labels):
            plt.plot(s, v, label='batch size ' + l)
            # plt.title(k)
            plt.xlabel('number of rounds')
            plt.ylabel(ylabel)
            # plt.xscale('log')
            plt.grid()
            plt.legend()
            plt.tight_layout()
            plt.savefig(f'init_PS_{name}_{query}.pdf')
```
